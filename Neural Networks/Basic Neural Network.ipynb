# Neural Networks 

also called Multi Layer perceptrons in scikit learn library are very popular when it comes to machine learning algorithms. The Backpropogation algorithms helps train the neural network. In this tutorial we apply Neural Networks to using scikit learn library on the MNIST handwriting dataset and check the accuracy. 

## Back Propogation



```python
#bring in the data sets 
import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
%matplotlib inline

print('done')
```

## Grab the data for the model

```python
#read in the data 
data=pd.read_csv('mnist.csv')
data 
```

## Split the data into train and test sets 

```python
df_x=data.iloc[:,1:]
df_y=data.iloc[:,0]

x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.2, random_state=4)

x_train
```

## Build out the neural network 

```python
nn=MLPClassifier(activation='logistic',solver='sgd',hidden_layer_sizes=(10,15),random_state=1)
nn
```

## Fit the model to the current data

```python
nn.fit(x_train,y_train)
```

## Test out the Predictions

activation logistic with hidden layer sizes-> 45,90 Gave 92 % accuracy
activation relu with hidden layer sizes-> 45,90 gave 89 % accuracy 
Test with different combinatations of learning rate, activation and other hyper parameters and mesure the accuracy

```python
pred=nn.predict(x_test)
pred
```

## Get the accuracy of the predictions train versus test

```python
a=y_test.values
count=0

for i in range(len(pred)):
    if pred[i]==a[i]:
        count+=1

count
```

```python
len(pred)
```

```python
print('current accuracy is')
count/len(pred)
```
